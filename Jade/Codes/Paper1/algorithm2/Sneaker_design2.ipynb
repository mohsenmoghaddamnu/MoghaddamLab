```python
def get_trees(file_name, s_location, p_location, m_location):
    # p_location is the position of the java pacakge of the jar, m_location is the nlp model of the 
    jar = p_location
    model = m_location
    parser = StanfordParser(model, jar, encoding = 'utf8')
    parser._classpath = tuple(find_jars_within_path(s_location))
    i  = 0
    for k, v in file_name.items():
        i += 1
        if i % 30 == 0:
            print("complete %s" % (i/len(file_name)) )
        tmp_tree = []
        for review in v:
            try:
                nima = list(parser.raw_parse(review))
                tmp_tree.append(nima)
            except:
                pass
        file_name[k] = tmp_tree
        
    return file_name
```
